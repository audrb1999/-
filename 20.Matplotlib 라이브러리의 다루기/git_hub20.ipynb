# 20.Matplotlib 라이브러리의 다루기

```python
# 선 그리기

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-9, 10) # -9 ~ 9 까지 1간격으로
y1 = x ** 2 # 그 값들을 제곱
plt.plot(
    x, y1,
    linestyle=":",
    marker="o",
    markersize=8,
    markerfacecolor="blue",# 중심 파란색
    markeredgecolor="red"# 테투리 빨간색
)
plt.show()
```

```python
# 누적 막대 그래프

import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(10) # 아래 막대, 길이 랜덤
y = np.random.rand(10) # 중간 막대
z = np.random.rand(10) # 위 막대
data = [x, y, z]# 리스트화
x_array = np.arange(10)# 0~9
for i in range(0, 3): # 누적 막대의 종류가 3개
  plt.bar(
    x_array, # 0부터 10까지의 X 위치에서
    data[i], # 각 높이(10개)만큼 쌓음, 랜덤하게
    bottom=np.sum(data[:i], axis=0)# 데이터가 쌓이는 과정 : 를 붙여줘야 쌓임
  )
plt.show()
```

```python
# 스캐터 그래프 > 특정한 공간에서 데이터의 분포를 표현할 때

import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(10)#10개의 랜덤한 x값
y = np.random.rand(10)
colors = np.random.randint(0, 100, 10)# 0~100까지 랜덤한 값으로 색 결정
sizes = np.pi * 1000 * np.random.rand(10) # 원 크기도 랜덤
plt.scatter(x, y, c=colors, s=sizes, alpha=0.7)# 알파는 투명도
plt.show()
```
