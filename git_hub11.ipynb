데이터 수집과 분석


문제를 직접 손으로 풀면서 문제 이미지 파일을 얻어보자

>데이터 수집

- 숫자들은 고정된 형태를 사용 > 데이터를 조금만 수집해도 됨

- 문제들은 갯수가 적으니 브라우저로 수집하자

- 가능한 12개의 문자 0~9, +, -, *를 수집해야함

- 대략적으로 최소한 5개 이상의 이미지가 필요!

>데이터 분석

- 색상 추출기를 활용하여 각 문자의 색상이 어떻게 구성됬는지 확인합니다.

- 색상 추출기로 #FFAAAA 와 HEX 값 같은 특징을 확인해야합니다.

- HEX값은 각각 FF(R), AA(G), AA(B)를 의미합니다.
즉 2개씩 묶어서 색깔로 생각하면 됩니다.
