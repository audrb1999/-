pandas란?

넘파이 라이브러리와 같이 자주 사용되는 라이브러리!

- 다수의 데이터를 효과적으로 처리하고 시각화하기 간단하게 보여주는 라이브러리

- 넘파이와 함께 사용되며 연계적인 기능을 제공

- 인덱스에 따라 데이터를 나열하므로 사전 자료형에 가깝습니다.

- 시리즈를 기본 자료형으로 사용합니다.

- 엑셀과 거의 똑같습니다.

정보를 담는 공간 = Frame이라고 합니다. > 2개 이상의 Series 가 있을 시에만

Frame은 Series(열벡터)로 구성이 됩니다.

Series는 인덱스와 값으로 구성되면 딕셔너리와 같은 형태를 띕니다.

{'name': '홍길동', 'age': 24}


```python
import pandas as pd

array = pd.Series(['사과', '바나나', '당근'], index = ['a', 'b', 'c'])

print(array)
print(array['a'])
```

데이터 프레임이란 다수의 시리즈를 모아 처리하기 위한 목적으로 사용됨!

표 형태로 데이터를 쉽게 출력 가능!

```python
import pandas as pd

word_dict = {'Apple':'사과','Banana':'바나나', 'Carrot':'당근'}

frequency_dict = {'Apple':3, 'Banana':5, 'Carrot':7}

word = pd.Series(word_dict)
frequency = pd.Series(frequency_dict)

summary = pd.DataFrame({'word': word, 'frequency':frequency})

print(summary)
```

데이터 프레임은 인덱스를 기준으로 각각의 데이터를 구성합니다. 

- 여기서 인덱스는    Apple    Banana  Carrot를 의미 두 시리즈 모두 동일한 인덱스로 묶여있음

## 판단스 연산

image.png

```python
# 동일한 인덱스끼리 곱해집니다. 
word_dict = {'Apple':'사과','Banana':'바나나', 'Carrot':'당근'}

frequency_dict = {'Apple':3, 'Banana':5, 'Carrot':7}

importance_dict = {'Apple':3, 'Banana':2, 'Carrot':1}


word = pd.Series(word_dict)
frequency = pd.Series(frequency_dict)
importance = pd.Series(importance_dict)

summary = pd.DataFrame({'word': word, 'frequency':frequency, 'importance':importance})

score = summary['frequency'] * summary['importance']

print(summary,'\n')
print(score)
```

## 데이터 프레임 슬라이싱

```python
# 이름(열)을 기준으로 슬라이싱
print(summary.loc['Banana':'Carrot', 'importance':]) # 열은 'importance'부터 끝까지

# 인덱스를 기준으로 슬라이싱
print(summary.iloc[1:3, 2:])
```

## 데이터 프레임 값 변경

```python
summary.loc['Apple', 'importance'] = 5 # 데이터 변경

summary.loc['Elderberry'] = ['엘더베리', 5, 3] # 인덱스 = 리스트는 행벡터로 추가

summary
```

## 판다스와 csv

```python
summary.to_csv('summary.csv', encoding = 'utf-8-sig') # csv로 변경, 저장, 한글데이터를 위해 인코딩
saved = pd.read_csv('summary.csv', index_col = 0)# 불러오기, 인덱스는 원래 그 상태로

print(saved)
```
