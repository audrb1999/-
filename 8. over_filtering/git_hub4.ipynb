```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('gray_image.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))#bgr을 rgb로 바꿔서 실행!
plt.title('Before', fontsize = 14)
plt.show()

size = 4
kernel = np.ones((size, size), np.float32) / size ** 2

"""  > 필터 모양  > 1/16 연산을 시행 > 주변에 있는 4x4 크기의 평균값으로 해당 픽셀을 바꿔주는 것!
1/16 1/16 1/16 1/16
1/16 1/16 1/16 1/16
1/16 1/16 1/16 1/16
1/16 1/16 1/16 1/16
"""

dst = cv2.filter2D(image, -1, kernel)# 만든 필터를 적용하는 함수!
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.title('After', fontsize = 14)
plt.show()#블러링 됨!
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('gray_image.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))#bgr을 rgb로 바꿔서 실행!
plt.title('Before', fontsize = 14)
plt.show()


dst = cv2.blur(image, (4, 4))# 이렇게해도 위랑 똑같은 이미지 나옴!
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.title('After', fontsize = 14)
plt.show()#블러링 됨!

# 블러링의 경우 커널 크기가 커질 수 록 전체 이미지의 평균값으로 대체되면서 흐려짐이 심해짐!
```

###Gaussian Blur

```python
import cv2
import matplotlib.pyplot as plt


image = cv2.imread('gray_image.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

#커널 사이즈 = 홀수 가우시안은 홀수 크기가 되야함!
dst = cv2.GaussianBlur(image, (5, 5), 0)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
