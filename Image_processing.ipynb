##***이미지 처리***

```python

#matplotlib 이용시 색깔 순서 RGB로 해야함!   >  open cv는 기본적으로 BGR
im_basic = cv2.imread('cat.jpg', cv2.IMREAD_COLOR)#경로는 이렇게 설정
plt.imshow(cv2.cvtColor(im_basic, cv2.COLOR_BGR2RGB))
plt.show()


im_gray = cv2.cvtColor(im_basic, cv2.COLOR_BGR2GRAY)#BGR형태를 GRAY로 바꾸겠다
plt.imshow(cv2.cvtColor(im_gray, cv2.COLOR_BGR2RGB))
plt.show()

```
